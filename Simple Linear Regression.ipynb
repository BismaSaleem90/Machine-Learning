```python
# to drop ka specific column
X=df.drop(columns=['Salary'])
Y=df['Salary']

# dropping x because we will predict it as output
# x= years of experience(input feature)
# y= salary (output feature/ prediction)
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)

# we are breaking our data into  2 parts training and testing data
# Variable	What it holds
# X_train	Input for training (years of experience to teach the model)
# X_test	Input for testing (years of experience for test questions)
# Y_train	Output for training (actual salaries used to learn)
# Y_test	Output for testing (actual salaries to check guesses)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
# call the model
reg = LinearRegression()
```

```python
reg
```

```python

# train the model
# with 2 parameters
# X_train	Input: experience
# Y_train	Output: salary
# fit()	Learn the pattern
reg.fit(X_train, Y_train)
```

```python
# Can you guess the salaries for X_TEST (YEARS OF EXPERIENCE)
y_pred= reg.predict(X_test)
```

```python
pd.DataFrame({'Y_test':Y_test, 'y_pred':y_pred, 'MeanAbsoluteError':(Y_test-y_pred)})
```

```python
# mean absolute error = actual - predicted
# mean square error =
# root mean square error (rmse)= 

from sklearn.metrics import mean_squared_error, r2_score
```

```python
# evaluate the model

# Metric	          What it Measures	                              Formula Units
# MAE	              Average error (regular)	                      Average of error
# MSE	              Average of squared errors	                      Average of (error)²	Squared units (e.g., 💵²)
# RMSE	              Like MSE but back to normal scale	              √MSE Same as output (e.g., salary 💵)


print('MSE = ', mean_squared_error(Y_test, y_pred))
print('R2 = ', r2_score(Y_test, y_pred))
print('RMSE = ', np.sqrt(mean_squared_error(Y_test, y_pred)))
```
